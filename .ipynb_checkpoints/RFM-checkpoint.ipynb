```python
import pandas as pd 
```

```python
orders = pd.read_csv("C:/Users/belou/OneDrive/Рабочий стол/пет проект/archive/orders.csv")
customers = pd.read_csv("C:/Users/belou/OneDrive/Рабочий стол/пет проект/archive/customers.csv")
order_items = pd.read_csv("C:/Users/belou/OneDrive/Рабочий стол/пет проект/archive/order_items.csv")
```

```python
df = orders.merge( #соединение таблиц 
    customers, 
    on='customer_id', 
    how='inner')
df = df.merge (
    order_items, 
    on='order_id', 
    how='inner')
df = df[df['order_status'] == 'delivered']
df.head()
```

```python
df['order_date'] = pd.to_datetime(df['order_purchase_timestamp'])#формат
df['order_date'].dtype #проверить формат 
```

```python
last_day = df['order_date'].max()
last_order = (
    df.groupby('customer_unique_id')['order_date']
    .max()
    .reset_index()
)
last_order['recency']=(last_day-last_order['order_date']).dt.days
last_order.head()
```

```python
frequency = (
    df.groupby('customer_unique_id')['order_id']
    .nunique()
    .reset_index()
    .rename(columns={'order_id': 'frequency'})
)
frequency.sort_values(by='frequency', ascending=False).head()
```

```python
monetary = (
    df.groupby('customer_unique_id')['price']
    .sum()
    .reset_index()
    .rename(columns={'price':'monetary'})
)
monetary.head()
```

```python
rfm = last_order.merge(
    frequency,
    on = 'customer_unique_id', 
    how='inner')
rfm = rfm.merge(
    monetary, 
    on = 'customer_unique_id', 
    how='inner')
rfm = rfm[['customer_unique_id', 'recency', 'frequency', 'monetary']]
rfm.head()
```

```python
rfm['R_score'] = pd.qcut(
    rfm['recency'], 
    q=3,
    labels=[3,2,1]
)
rfm['M_score'] = pd.qcut(
    rfm['monetary'], 
    q=3, 
    labels=[1,2,3]
)
```

```python
rfm['frequency'].value_counts().sort_index()
```

```python
rfm['F_score'] = pd.cut(
    rfm['frequency'], 
    bins=[0,1,2,rfm['frequency'].max()], 
    labels=[1,2,3],
    right=True
)
rfm['F_score'].value_counts().sort_index()
```

```python
rfm['rfm_code'] = (
    rfm['R_score'].astype(str)  
    + rfm['F_score'].astype(str)
    + rfm['M_score'].astype(str)
)
rfm.head()
```

```python
rfm['rfm_code'].value_counts()
```

```python
def rfm_segment(row):#деф - слово, которое обозначает начало функции, (роу) - аргумент одна строка 
    if row['R_score']==3 and row['F_score']==3 and row['M_score']==3: return 'VIP' #самые крутые 
    elif row['R_score']>1 and row['F_score']>1 and row['M_score']>1: return 'Loyal' #средне или недавно сделали покупку, 2 или больше заказов 
    elif row['R_score']==3 and row['F_score']==1: return 'New customer' #недавно сделали первый заказ 
    elif row['R_score']==1 and row['F_score']>1: return 'At risk' #давно не покупали, но раньше были активны 
    elif row['R_score']==1 and row['F_score']==1: return 'Lost' #сделали 1 заказ и давно 
    else: return 'Others'
rfm['segment'] = rfm.apply(rfm_segment, axis=1) #функция, чтобы применить функцию rfm_segment к каждой строке, axis=1 → Функция получает строку, а не столбец.
rfm=rfm[['customer_unique_id', 'recency', 'frequency', 'monetary', 'R_score', 'F_score', 'M_score', 'rfm_code', 'segment']] 
rfm.head()
```
